```python
import pandas as pd

df = pd.read_csv("Hospital_Analytics_Dataset.csv")

print(df.head())
```

```python
 #Revenue by Department
dept_rev = df.groupby('Department')['Bill_Amount'].sum()

print(dept_rev)
```

```python
#Top Doctors
top_d = df.groupby('Doctor')['Patient_ID'].count()
print(top_d)
```

```python
#Monthly Revenue
df['Visit_Date'] = pd.to_datetime(df['Visit_Date'])

df['Month'] = df['Visit_Date'].dt.month

monthly = df.groupby('Month')['Bill_Amount'].sum()

print(monthly)
```
